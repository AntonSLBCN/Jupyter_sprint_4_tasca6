```python
#Gràfic directe
df_tsc6.groupby('UniqueCarrier')['UniqueCarrier'].count().plot(kind= 'bar')
#Exercici 2
plt.savefig('Fig1.jpg')
```

```python
#Histogram
df_tsc6['UniqueCarrier'].hist()
#Exercici 2
plt.savefig('Fig2.jpg')
```

```python
#metode countplot de seaborn per comptar variables categògiriques.
sns.countplot(x="UniqueCarrier", data = df_tsc6, palette="Greens_d" )
#Exercici 2
plt.savefig('Fig3.jpg')
```

* <span style=color:blue>Una variable numèrica (ArrDelay)</span>

```python
grafico1, ax = plt.subplots()
xg2=df_tsc6['ArrDelay']
plt.xlabel('Llegadas con retraso')
ax.plot(xg2, 'o:m')
plt.show()
#Exercici 2
grafico1.savefig('C:\\Users\\Anton SL\\OneDrive\\Documentos\\IT_Academics\\Data_Science\\Jupyter_sprint_4_tasca6\\Fig4.png', bbox_inches='tight')
#plt.savefig('foo.png', bbox_inches='tight')
```

```python
df_tsc6['ArrDelay'].hist(bins=100 , width=0.5)
#Exercici 2
plt.savefig('Fig5.jpg')
```

* <span style=color:blue>Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)</span>

```python
sns.barplot(x= 'UniqueCarrier', y= 'ArrDelay', data=df_tsc6, orient='v')
#Exercici 2
plt.savefig('Fig6A.jpg')
```

* <span style=color:blue>Dues variables numèriques (ArrDelay i DepDelay)</span>

```python
f = plt.figure()
f1 = f.add_subplot(1,1,1)
f1.plot(df_tsc6['DepDelay'].cumsum(), 'g--',label='Delayed Departures')
f1.plot(df_tsc6['ArrDelay'].cumsum(), 'r--',label='Delayed Departures')
f1.legend(loc='best')
#Exercici 2
plt.savefig('Fig7.jpg')
```

```python
f = plt.figure()
f2 = f.add_subplot(1,1,1)
f2.plot(df_tsc6['CRSDepTime'].cumsum(), 'o',label='Delayed Departures')
f2.plot(df_tsc6['CRSArrTime'].cumsum(), 'r--',label='Delayed Arrives')
f2.legend(loc='best')
#Exercici 2
plt.savefig('Fig8.jpg')
```

```python
# mismo grafico con seaborn
sns.scatterplot(x = df_tsc6['ArrDelay'], y = df_tsc6['DepDelay'])
#Exercici 2
plt.savefig('Fig9.jpg')
```

* <span style=color:blue>Tres variables (ArrDelay, DepDelay i UniqueCarrier)</span>

```python
Grafic, ax = plt.subplots()
#f = plt.figure()
#f2 = f.add_subplot(1,1,1)
x = df_tsc6['UniqueCarrier']
y1 = df_tsc6['ArrDelay']
y2 = df_tsc6 ['DepDelay']
ax.scatter(x,y1, c= 'red')
ax.plot(x,y2)
plt.show()
#Exercici 2
Grafic.savefig('C:\\Users\\Anton SL\\OneDrive\\Documentos\\IT_Academics\\Data_Science\\Jupyter_sprint_4_tasca6\\Fig10.jpg')
```

```python
df_tsc7 = df_tsc6[['ArrDelay', 'DepDelay' , 'UniqueCarrier']]
df_tsc7.groupby(by =['UniqueCarrier']).mean().plot.bar()
#Exercici 2
plt.savefig('Fig11.jpg')
```

* <span style=color:blue>Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier</span>

```python
df_tsc7 = df_tsc6[['ArrDelay', 'DepDelay' ,'AirTime' ,'UniqueCarrier']]
df_tsc7.groupby(by =['UniqueCarrier']).mean().plot.bar()
#Exercici 2
plt.savefig('Fig12.jpg')
```

```python
sns.pairplot(data = df_tsc7, hue='UniqueCarrier' )
#Exercici 2
plt.savefig('Fig13.jpg')
```
